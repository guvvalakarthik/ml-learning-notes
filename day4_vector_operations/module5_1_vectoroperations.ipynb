Cross product

```python
vector_1=np.asanyarray([1,2,3])
vector_2=np.asarray([3,2,1])
```

```python
cross_1=np.cross(vector_1,vector_2)
print(cross_1)
```

```python
vector_1=np.asanyarray([1,2,-1])
vector_2=np.asarray([3,2,-4])
```

```python
cross_2=np.cross(vector_1,vector_2)
print(cross_2)
```

Projection of a vector


```python
vector_a=np.asarray([1,2,3])
vector_v=np.asarray([4,5,6])
```

```python
magnitude_of_result=np.sqrt(np.sum(vector_v**2))
print(magnitude_of_result)
```

```python
dot_product=np.dot(vector_a,vector_v)
print(dot_product)
```

```python
projection=(dot_product/magnitude_of_result**2)*vector_v
print(projection)
```
